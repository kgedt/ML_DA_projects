Датасет с числом самоубийств в Казахстане в разрезе по годам(1990-2015), полу, возрасту и т. д.

Построим графики по годам для подростков возраста 15-24 года(отдельно для мужчин и женщин)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Считываем csv файл, удаляем пустые строки и отбираем данные исключительно для Казахстана

```python
df = pd.read_csv("./suicides.csv")
df.dropna()
df_kz = df[df['country'] == 'Kazakhstan']
df_kz.head(10)
```

Фильтруем датасет по возрасту от 15 до 24 лет и разделяем по полу

```python
df_kz = df_kz[df_kz['age'] == '15-24 years']

males = df_kz[df_kz['sex'] == 'male']
females = df_kz[df_kz['sex'] == 'female']

x_males = males['year']
y_males = males['suicides_no']

x_females = females['year']
y_females = females['suicides_no']
```

Визуализируем. График числа мужчин покрашен синим, женщин — розовым

```python
# fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15, 6), sharey=True)
fig, ax = plt.subplots(figsize=(10, 7))
ax.grid()
ax.plot(x_males, y_males, 'blue', 
        x_females, y_females, 'violet')

#axes[0].set_xlabel('Среди мужчин')
#axes[0].set_ylabel('Число самоубийств')
#axes[0].grid()

#axes[1].set_xlabel('Среди девушек')
#axes[1].set_ylabel('Число самоубийств')
#axes[1].grid()

#ax1.plot(x_males, y_males, color='blue')
#ax2.plot(x_females, y_females, color='pink')
```
