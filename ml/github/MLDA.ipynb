# Data Preparation

```python
df.info()
```

## Tackle the missing values

```python
df.isnull().sum()
```

There seems to be no null values, however, rerun_ID is NaN, so we need to drop it

```python
df = df.drop('rerun_ID', axis = 1)
```

```python
df.head()
```

```python
df['class'].value_counts()
```

```python
sb.countplot(x = 'class', data = df, hue = 'class')
```

'class' is categorical data. We need to transform it into numberical for the models to work.

```python
df['class'] = [0 if i == 'GALAXY' else 1 if i == 'STAR' else 2 for i in df['class']]
```

```python
sb.countplot(x = 'class', data = df, hue = 'class')
```

The 'class' data is now numerical. However, we can see that there is much more Galaxy data than everything else. If we run model like this, there would be quantity bias towards Galaxy class. The data needs to be balanced. It is shown on previous plots that there is a lot of outliers, so they has to be removed, so the balancing will be more accurate. 

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor()
y_pred = lof.fit_predict(df)
```

```python
x_score = lof.negative_outlier_factor_
outlier_score = pd.DataFrame()
outlier_score["score"] = x_score

#threshold
threshold = -1.5                                            
filtre = outlier_score["score"] < threshold
outlier_index = outlier_score[filtre].index.tolist()
```

```python
len(outlier_index)
```

```python
df.drop(outlier_index, inplace=True)
df.info()
```

There are still some columns that are not relevent to the prediction. Columns like IDs and columns with low correlation to the 'class' attribute need to be removed.

```python
df.corr()['class'].sort_values()
```

```python
df = df.drop(['obj_ID', 'run_ID', 'field_ID', 'fiber_ID', 'alpha', 'delta', 'cam_col'], axis = 1)
```

```python
df.info()
```

Now, when we removed outliers and irrelevant columns, we can balance 'class' attribute with SMOTE algoritm.

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

x = df.drop(['class'], axis = 1)
y = df.loc[:,'class'].values

smt = SMOTE(random_state=42)
print(f'Original dataset shape {Counter(y)}')
x, y = smt.fit_resample(x, y)
print(f'Balanced dataset shape {Counter(y)}' )
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33,
                                                    random_state = 42)

print(f'x_train: {x_train.shape}')
print(f'x_test: {x_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

# Model training

## Support Vector Machine model

```python
from sklearn import svm

svm_clf = svm.SVC(kernel='rbf', C=10, random_state=42)
svm_clf.fit(x_train,y_train)
predicted = svm_clf.predict(x_test)
score = svm_clf.score(x_test, y_test)
svm_score_ = np.mean(score)

print('Accuracy : %.3f' % (svm_score_))
```

## Classification metrics

```python
from sklearn.metrics import confusion_matrix, classification_report
from yellowbrick.classifier import ConfusionMatrix

print(classification_report(y_test, predicted))
```

## Confusion matrix

```python
classes = ['GALAXY','STAR','QSO']
svm_confmatx = ConfusionMatrix(svm_clf, classes=classes)
svm_confmatx.fit(x_train, y_train)
svm_confmatx.score(x_test, y_test)
svm_confmatx.show()
```

## K-Nearest Neighbors model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_predict_knn = knn.predict(x_test)
score = knn.score(x_test, y_test)
knn_score_ = np.mean(score)
print('Accuracy : %.3f' % (knn_score_))
```

## Classification metrics

```python
print(classification_report(y_test, y_predict_knn))
```

## Confusion Matrix

```python
knn_cm = ConfusionMatrix(knn, classes=classes)
knn_cm.fit(x_train, y_train)
knn_cm.score(x_test, y_test)
knn_cm.show()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=300)
logreg.fit(x_train,y_train)
y_predict_logreg = logreg.predict(x_test)
score = logreg.score(x_test, y_test)
logreg_score_ = np.mean(score)
print('Accuracy : %.3f' % (logreg_score_))
```

## Classification metrics

```python
print(classification_report(y_test, y_predict_logreg))
```

## Confusion Matrix

```python
logreg_cm = ConfusionMatrix(logreg, classes=classes)
logreg_cm.fit(x_train, y_train)
logreg_cm.score(x_test, y_test)
logreg_cm.show()
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

randfor = RandomForestClassifier(random_state=42)
randfor.fit(x_train,y_train)
y_predict = randfor.predict(x_test)
score = randfor.score(x_test, y_test)
randfor_score_ = np.mean(score)
print('Accuracy : %.3f' % (randfor_score_))
```

## Classification metrics

```python
print(classification_report(y_test, y_predict))
```

## Confusion Matrix

```python
randfor_cm = ConfusionMatrix(randfor, classes=classes)
randfor_cm.fit(x_train, y_train)
randfor_cm.score(x_test, y_test)
randfor_cm.show()
```
