## (p. 27, ex. 7) Draw a relative frequency table for the data of Prob. 1. Plot these relative frequencies in a line graph.
data: [5, 13, 9, 12, 7, 4, 8, 6, 6, 10, 7, 11, 10, 8, 15, 8, 6, 9, 12, 10, 7, 11, 10, 8, 12, 9, 7, 10, 7, 8]

```python
data = np.array([5, 13, 9, 12, 7, 4, 8, 6, 6, 10, 7, 11, 10, 8, 15, 8, 6, 9, 12, 10, 7, 11, 10, 8, 12, 9, 7, 10, 7, 8])
first_unique = np.array(list(set(data)), dtype=int)
first_data_min = first_unique[0]
first_data_max = first_unique[-1]
first_frequency = np.zeros(first_data_max - first_data_min + 1, dtype=int)
for i in data:
    first_frequency[i-first_data_min] += 1
zeros_ind = list()
for i in range(len(first_frequency)):
    if 0 == first_frequency[i]:
        zeros_ind.append(i)
first_frequency = np.delete(first_frequency, zeros_ind)
```

```python
first_frequency_table = pd.DataFrame({'family_size' : first_unique, 'frequency' : first_frequency})
first_frequency_table
```

```python
fig, axs = plt.subplots(figsize=(15, 7))
axs.grid()
axs.plot(first_unique, first_frequency, marker='o', color='b')
axs.set_xlabel('Family size')
axs.set_ylabel('Frequency')
axs.locator_params(axis='x', nbins=2*len(first_unique))
```

## Boxplot: median, quantiles, IQR
data = [34, 45, 73, 36, 50, 52, 16, 27, 30, 60, 24, 27, 28, 29, 46, 71, 72]

```python
data = np.array([34, 45, 73, 36, 50, 52, 16, 27, 30, 60, 24, 27, 28, 29, 46, 71, 72])
outliers = np.array([])
median, Q1, Q3, IQR = [float()] * 4
data.sort()
median = np.quantile(data, .5)
Q1 = np.quantile(data, .25)
Q3 = np.quantile(data, .75)
IQR = Q3 - Q1
min_max = [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
for value in data:
    if min_max[0] > value > min_max[1]:
        np.append(outliers, value)
```

```python
print(f'''sorted data={data}
median={median}
Q1={Q1}
Q3={Q3}
IQR={IQR}
min={min_max[0]}
max={min_max[1]}
outliers={outliers}''')

width = 3
fig = plt.figure(figsize=(10, 7), facecolor='gray')
ax = fig.add_subplot(facecolor='beige', xlabel='DATA RANGE')
bp = ax.boxplot(data, vert=False)

bp['boxes'][0].set(color='orange',
                  linewidth=width)
for whisker in bp['whiskers']:
    whisker.set(color='#ff0000',
                linewidth=width,
                linestyle='-.')
    
for cap in bp['caps']:
    cap.set(color='violet',
            linewidth=width)
```
